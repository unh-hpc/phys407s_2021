## Problem 2

Given: $m = 2.00 \times 10^6$ kg, $F_{thrust} = 3.50 \times 10^7 N$.

### Part a)

As usual, let's draw a FBD.

![figure](hw5_2.png)

It's a 1-d problem, I'm choosing the up direction to be positive.

\begin{align}
ma &= F_{thrust} - mg\\
a &= \frac{f_{thrust}}{m} - g\\
\end{align}


```python
m = 2e6
F_thrust = 3.5e7
a = F_thrust/m - g
print(f"a = {a:.2f}")
```

### Part b)

Use some kinematics -- since $a$ is constant, $a = \frac{\Delta v}{\Delta t}$, so $\Delta t = \frac{v_f - v_i}{a}$, where $v_i = 0$, $v_f = 120$ km/h.

$v_f = 120 \frac{km}{h} \frac{1000 m}{km} \frac{h}{3600 s}$

```python
v_f = 120 * 1000 / 3600
v_i = 0
dt = (v_f - v_i) / a
print(f"dt = {dt:.2f} seconds")
```
