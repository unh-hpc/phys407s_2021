Just generic python setup first, including some helper functions to help with degree-radians conversion, and solving quadratic equations

```python
import numpy as np
from numpy import sin, cos, tan, arctan2, sqrt
```

```python
def rad(th):
    return th * np.pi/180

def deg(th):
    return th * 180/np.pi

def quadratic(a, b, c):
    return ((-b + sqrt(b**2 - 4*a*c)) / (2*a),
            (-b - sqrt(b**2 - 4*a*c)) / (2*a))
g = 9.8
```

## Problem 3

Given: angle $\theta = 10^\circ$, coefficient of friction $\mu = 0.100$.

As usual, let's draw a FBD.

![figure](hw5_3.png)

To find the acceleration of the package (in the positive $x$ direction), let's use Newton's 2nd Law -- and we'll need it in the $y$ direction, too, to find the normal force.

\begin{align}
ma &= F_{net,x} = F_{G,x} + F_{f, x} = mg \sin\theta - \mu F_N\\
0 &= F_{net,y} = F_N - F_{G,y} = F_N - mg \cos\theta
\end{align}

From the second equation, we find $F_N = mg\cos\theta$, plugging that into the first equation we can solve for $a$:

$$
a = (\sin\theta - \mu\cos\theta)g
$$

### Part a)

Let's plug things in...

```python
theta = rad(10)
mu = 0.1
a = (sin(theta) - mu*cos(theta))*g
print(f"a = {a:.2f} m/s^2")
```

### Part b)

$\mu$ is the same, we're looking for $\theta$ now, knowing that at constant speed: $a = 0$.

\begin{align}
0 &= (\sin\theta - \mu\cos\theta)g\\
\sin\theta &= \mu\cos\theta\\
\tan\theta &= \mu\\
\theta &= \tan^{-1}\mu
\end{align}

```python
theta = deg(np.arctan(mu))
print(f"theta = {theta:.2f} degrees")
```
